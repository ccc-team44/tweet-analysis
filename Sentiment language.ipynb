```python
def find_address(address):
    address = address.split(', ')
    if len(address) <= 1:
        return address[0]
    else:
        return address[0]

        
def find_location(address):
    address = address.split(', ')
    if len(address) <= 1:
        return ''
    else:
        return address[1]
```

```python
import json
with open('./au.json','r') as f:
    audata = json.load(f)
```

```python
state = {}
for info in audata:
    admin = info['admin']
    if admin in state.keys():
        state[admin].append(info['city'])
    else:
        state[admin] = []
        state[admin].append(info['city'])
```

```python
def find_loc(address):
    for state_name in state.keys():
        if address.find(state_name)!= -1:
            return state_name
        
        for city_name in state[state_name]:
            if address.find(city_name)!= -1:
                return state_name
        
    return None
```

```python
df['location'] = df['city'].apply(find_loc)
```

```python
def lga_code_to_region(code):
    if str(code).startswith('1'):
        return 'New South Wales'
    elif str(code).startswith('2'):
        return 'Victoria'
    elif str(code).startswith('3'):
        return 'Queensland'
    elif str(code).startswith('4'):
        return 'South Australia'
    elif str(code).startswith('5'):
        return 'Western Australia'
    elif str(code).startswith('6'):
        return 'Tasmania'
    else:
        return 'Northern Territory'
```

```python
aurin_data['index'] = aurin_data['index'].apply(lga_code_to_region)
```

```python
aurin = aurin_data.groupby(['index']).sum()
```
